```python
def loadAudio(url, start, length):

    rate, data = wavfile.read(url)
    if len(data.shape) > 1:
        data = data.sum(axis=1)
    data = (1.0 * data / abs(data).max()).astype(np.float32)
    return rate, data[rate*start+np.arange(rate*length)]
```
