## 1. Programmatically generate a sinusoidal wave, running at 500hz, an amplitude of 5 volts, with a sampling rate of 100, starting at time 0 with 0 offset. Store the sine wave in an array.

```python
import numpy as np
import matplotlib.pyplot as plt

startTime = 0
endTime = 1
sampleRate = 100
time = np.arange(startTime, endTime,1/sampleRate)

theta = 0
frequency = 500
amplitude = 5
sinewave = amplitude * np.sin(2 * np.pi * frequency * time + theta)
```

## 4. Create an 10-element array containing the following kernel function with values:

## 8. Generate another kernel arrays using the scipy wavelet library and convolve with the sine wave. This time use the following:
        Daubechie wavelet 
        signal.wavelets.daub(8)
        Morlet wavelet 
        signal.wavelets.morlet2(20,4,2)
        Ricker wavelet
        signal.wavelets.ricker(20,1)


```python
from scipy import signal

daubechieWavelet  = signal.wavelets.daub(8)
morletWavelet = signal.wavelets.morlet2(20,4,2)
rickerWavelet = signal.wavelets.ricker(20,1)
```

```python
daubechieWaveletTitle = "Daubechie_wavelet_Plot"
plt.figure(1,figsize=(18,10))
plt.title(daubechieWaveletTitle)
plt.plot(daubechieWavelet)
plt.savefig(daubechieWaveletTitle+'.png', facecolor='white', transparent=False)

daubConv = np.convolve(sinewave, daubechieWavelet)
convTitle = "Convolved_Daubechie_wavelet_and_sinewave"
plt.figure(2,figsize=(18,10))
plt.title(convTitle)
plt.plot(daubConv)
plt.savefig(convTitle+'.png',facecolor='white', transparent=False)
```

```python
morletWaveletTitle = "Morlet_wavelet_Plot"
plt.figure(1,figsize=(18,10))
plt.title(morletWaveletTitle)
plt.plot(morletWavelet)
plt.savefig(morletWaveletTitle+'.png', facecolor='white', transparent=False)

morletConv = np.convolve(sinewave, morletWavelet)
convTitle = "Convolved_Morlet_wavelet_and_sinewave"
plt.figure(2,figsize=(18,10))
plt.title(convTitle)
plt.plot(morletConv)
plt.savefig(morletWaveletTitle+'.png',facecolor='white', transparent=False)
```

```python
rickerWaveletTitle = "Ricker_wavelet_Plot"
plt.figure(1,figsize=(18,10))
plt.title(rickerWaveletTitle)
plt.plot(rickerWavelet)
plt.savefig(rickerWaveletTitle+'.png', facecolor='white', transparent=False)

rickerConv = np.convolve(sinewave, rickerWavelet)
convTitle = "Convolved_Ricker_wavelet_and_sinewave"
plt.figure(2,figsize=(18,10))
plt.title(convTitle)
plt.plot(rickerConv)
plt.savefig(rickerWaveletTitle+'.png',facecolor='white', transparent=False)
```

## 8. Compare and analyses the different effects of the kernels to the original signal and each other.


```python
plt.figure(1,figsize=(18,10))
plt.title("Comparing daub convoled and sinewave")
plt.plot(sinewave, label = "Original")
plt.plot(daubConv, label = "Convolved")
plt.legend()
plt.savefig('Comparing_daub_convoled_and_sinewave.png', facecolor='white', transparent=False)

plt.figure(2,figsize=(18,10))
plt.title("Comparing morlet convoled and sinewave")
plt.plot(sinewave, label = "Original")
plt.plot(morletConv, label = "Convolved")
plt.legend()
plt.savefig('Comparing_morlet_convoled_and_sinewave.png', facecolor='white', transparent=False)

plt.figure(3,figsize=(18,10))
plt.title("Comparing ricker convoled and sinewave")
plt.plot(sinewave, label = "Original")
plt.plot(rickerConv, label = "Convolved")
plt.legend()
plt.savefig('Comparing_ricker_convoled_and_sinewave.png', facecolor='white', transparent=False)

plt.figure(4,figsize=(18,10))
plt.title("Comparing 3 convoled wavelets and sinewave")
plt.plot(sinewave, label = "Original")
plt.plot(daubConv, label = "Daub Convolved")
plt.plot(morletConv, label = "Morlet Convolved")
plt.plot(rickerConv, label = "Ricker Convolved")
plt.legend()
plt.savefig('Comparing_3_convoled_wavelets_and_sinewave.png', facecolor='white', transparent=False)
```

## 9. Apply the kernels A-G and the three(3) wavelets to the mean and standard deviation summaries from activity 1. 


```python
import pandas as pd

pathtoMeanStd = r'D:\Users\Ervie John\Documents\4th year 1st sem\Digital Signal Processing and Application\Module 1\CPE-027-CPE41S2_Lab1-Group3\Compiled_Csv_Files\AllCsvFiles.csv'
MeanStd = pd.read_csv(pathtoMeanStd)

Conv_Mean = []
A_G_list = [A, B, C, D, E, F, G]
for i in A_G_list: # A - G
    Conv_Mean.append(np.convolve(MeanStd['Mean'], i))

Conv_Mean.append(np.convolve(MeanStd['Mean'], daubechieWavelet))
Conv_Mean.append(np.convolve(MeanStd['Mean'], morletWavelet))
Conv_Mean.append(np.convolve(MeanStd['Mean'], rickerWavelet))

Conv_STD = []
for j in A_G_list: # A - G
    Conv_STD.append(np.convolve(MeanStd['Standard Deviation'], j))

Conv_STD.append(np.convolve(MeanStd['Standard Deviation'], daubechieWavelet))
Conv_STD.append(np.convolve(MeanStd['Standard Deviation'], morletWavelet))
Conv_STD.append(np.convolve(MeanStd['Standard Deviation'], rickerWavelet))
```

## 10. Cite and record observations.

```python
plt.figure(1,figsize=(18,10))
plt.title('Original Mean Convoled by A-G')
plt.plot(MeanStd['Mean'], label="Mean")
for k in range(0,7): # A-G
    plt.plot(Conv_Mean[k])
plt.legend(('Original Mean', 'A', 'B','C','D','E','F','G'))
plt.savefig('Original_Mean_Convoled_by_A-G.png', facecolor='white', transparent=False)
plt.show()

plt.figure(2,figsize=(18,10))
plt.title('Original Mean Convoled by 3 wavelets (Daubechie, Morlet, and Ricker)')
plt.plot(MeanStd['Mean'], label="Mean")
for l in range(7,10):
    plt.plot(Conv_Mean[l])
plt.legend(('Original Mean', 'Daubechie', 'Morlet', 'Ricker'))
plt.savefig('Original_Mean_Convoled_by_3_wavelets_(Daubechie,_Morlet,_and_Ricker).png', facecolor='white', transparent=False)
plt.show()
```

```python
plt.figure(1,figsize=(18,10))
plt.title('Original Standard Deviation Convoled by A-G')
plt.plot(MeanStd['Standard Deviation'], label="Mean")
for k in range(0,7): # A-G
    plt.plot(Conv_STD[k])
plt.legend(('Original STD', 'A', 'B','C','D','E','F','G'))
plt.savefig('Original_Standard_Deviation_Convoled_by_A-G.png', facecolor='white', transparent=False)
plt.show()

plt.figure(2,figsize=(18,10))
plt.title('Original Standard Deviation Convoled by 3 wavelets (Daubechie, Morlet, and Ricker)')
plt.plot(MeanStd['Standard Deviation'], label="Mean")
for l in range(7,10):
    plt.plot(Conv_STD[l])
plt.legend(('Original STD', 'Daubechie', 'Morlet', 'Ricker'))
plt.savefig('Original_Standard_Deviation_Convoled_by_3_wavelets_(Daubechie,_Morlet,_and_Ricker).png', facecolor='white', transparent=False)
plt.show()
```
